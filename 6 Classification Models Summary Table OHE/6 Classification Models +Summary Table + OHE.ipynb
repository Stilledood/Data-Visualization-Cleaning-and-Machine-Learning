### Import Libraries

```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import precision_score,recall_score,f1_score
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report,ConfusionMatrixDisplay
```


### Using One Hot Encoder

```python
one_hot_encoded_data=pd.get_dummies(students_dataset,columns=['Gender', 'Age', 'Education Level', 'Institution Type', 'IT Student',
       'Location', 'Load-shedding', 'Financial Condition', 'Internet Type',
       'Network Type', 'Class Duration', 'Self Lms', 'Device'])
```

```python
one_hot_encoded_data.head()
```

```python
one_hot_encoded_data['Adaptivity Level']=one_hot_encoded_data['Adaptivity Level'].map({'Moderate':2,'Low':1,'High':0})
```

```python
one_hot_encoded_data.head()
```

```python
X=one_hot_encoded_data.drop(['Adaptivity Level'],axis=1)
```

```python
X.head()
```

```python
y=one_hot_encoded_data[['Adaptivity Level']]
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.3)
```

### Logistic Multiclass

```python
lr=LogisticRegression(random_state=42,multi_class='multinomial',solver='lbfgs',max_iter=500,penalty='l2',warm_start=True)
```

```python
lr.fit(X_train,y_train.values.ravel())
```

```python
lr.get_params()
```

```python
pred_lr=lr.predict(X_test)
```

```python
print(classification_report(y_test,pred_lr,zero_division=1))
```

```python
cm=confusion_matrix(pred_lr,y_test)
disp=ConfusionMatrixDisplay(cm,display_labels=['High','Low','Moderate'])
disp.plot()
plt.title('Confusion Matrix')
```

## 5.Decision Tree

```python
clf=tree.DecisionTreeClassifier()
```

```python
params={
    'max_depth':range(1,20,2),
    'min_samples_leaf':range(1,100,5),
    'min_samples_split':range(2,10),
    'criterion':['gini','entropy'],
    'splitter':['best','random'],
    'max_features':['auto']
}
```

```python
grid_search=GridSearchCV(estimator=clf,param_grid=params,cv=4,n_jobs=-1,verbose=1,scoring='accuracy',error_score='raise')
```

```python
grid_search.fit(X_train,y_train.values.ravel())
```

```python
score_df=pd.DataFrame(grid_search.cv_results_)
```

```python
score_df.head()
```

```python
dt_best=grid_search.best_estimator_
```

```python
type(dt_best)
```

```python
def evaluate_model(dt_model):
    print("Train accuracy:",accuracy_score(y_train,dt_model.predict(X_train)))
    print(confusion_matrix(y_train,dt_model.predict(X_train)))
    print('Test accuracy:',accuracy_score(y_test,dt_model.predict(X_test)))
    print(confusion_matrix(y_test,dt_model.predict(X_test)))
```

```python
evaluate_model(dt_best)
```

```python
predictions=dt_best.predict(X_test)
```

```python
print(classification_report(y_test,predictions,zero_division=1))
```
